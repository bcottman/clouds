## "Week 6 - Excercise 12

```python
from diagrams.aws.integration import SNS
from diagrams.aws.integration import SQS 
from diagrams.onprem.client import Client

with Diagram("Week 6 - Excercise 12 - Archecture", direction='LR') as diag: 
    with Cluster("EC2 Cluster"):
        svc_group = [EC2("EC2 1"),
                    EC2("EC2  2")]
    vpc = VPC("vpc")
    rek = Rekognition("Rekognition")
    database = Mysql("RDS mySQL")
    s3 = SimpleStorageServiceS3("S3")
    fnet = CloudFront("https")
    load_balancer = ELB("Elastic Load Balancer")
    accounts = Cognito("Accounts")
    profiler = XRay("Profiler")
    lambdax = Lambda("microservive")
    sns = SNS("Topic: LoadImage")
    sqs = SQS("Print: Image")
    client = Client("Printer")
    

    fnet >> load_balancer >> svc_group
    svc_group >> vpc >> accounts
    svc_group >> vpc >> s3 >> sns >> lambdax
    lambdax >> rek
    sns >> lambdax
    sns << sqs >> client
    lambdax >> database >> profiler
diag

```
